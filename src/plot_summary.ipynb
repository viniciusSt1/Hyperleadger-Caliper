```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Carrega os CSVs
df = pd.read_csv("./reports/reportStatus/reportStatus_summary_metrics.csv")
df2 = pd.read_csv("./reports/statusReports/statusReports_summary_metrics.csv")
df3 = pd.read_csv("./reports/getLatestStatus/getLatestStatus_summary_metrics.csv")

# Define a ordem dos TPS
ordered_tps = sorted(set(df['ConfiguredTPS'].unique()).union(df2['ConfiguredTPS'].unique()).union(df3['ConfiguredTPS'].unique()))

# ===================== THROUGHPUT =====================
plt.figure(figsize=(8, 6))

# Boxplot TPS 1 - reportStatus
throughput_data_1 = [df[df['ConfiguredTPS'] == tps]['Throughput'] for tps in ordered_tps]
plt.boxplot(throughput_data_1, positions=np.arange(len(ordered_tps))-0.25, widths=0.10,
            patch_artist=True, boxprops=dict(facecolor='lightcoral', alpha=0.6),
            medianprops=dict(color='darkred', linewidth=2), showfliers=False)
mean_throughput_1 = [np.mean(data) for data in throughput_data_1]
plt.plot(np.arange(len(ordered_tps))-0.25, mean_throughput_1, linestyle='-', marker='o', color='red', label='reportStatus')

# Boxplot TPS 2 - statusReports
throughput_data_2 = [df2[df2['ConfiguredTPS'] == tps]['Throughput'] for tps in ordered_tps]
plt.boxplot(throughput_data_2, positions=np.arange(len(ordered_tps)), widths=0.10,
            patch_artist=True, boxprops=dict(facecolor='skyblue', alpha=0.6),
            medianprops=dict(color='darkblue', linewidth=2), showfliers=False)
mean_throughput_2 = [np.mean(data) if len(data) > 0 else np.nan for data in throughput_data_2]
plt.plot(np.arange(len(ordered_tps)), mean_throughput_2, linestyle='--', marker='s', color='blue', label='statusReports')

# Boxplot TPS 3 - getLatestStatus
throughput_data_3 = [df3[df3['ConfiguredTPS'] == tps]['Throughput'] for tps in ordered_tps]
plt.boxplot(throughput_data_3, positions=np.arange(len(ordered_tps))+0.25, widths=0.10,
            patch_artist=True, boxprops=dict(facecolor='mediumseagreen', alpha=0.6),
            medianprops=dict(color='darkgreen', linewidth=2), showfliers=False)
mean_throughput_3 = [np.mean(data) if len(data) > 0 else np.nan for data in throughput_data_3]
plt.plot(np.arange(len(ordered_tps))+0.25, mean_throughput_3, linestyle='-.', marker='^', color='green', label='getLatestStatus')

plt.xticks(ticks=np.arange(len(ordered_tps)), labels=ordered_tps)
plt.title('Throughput por TPS Configurado', fontsize=14)
plt.xlabel('TPS Configurado', fontsize=12)
plt.ylabel('Throughput (TPS)', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.4)
plt.legend()
plt.tight_layout()
plt.savefig("grafico_thoughput_tps.png", dpi=300)
plt.show()


# ===================== LATNCIA =====================
plt.figure(figsize=(8, 6))

# Boxplot TPS 1 - reportStatus
latency_data_1 = [df[df['ConfiguredTPS'] == tps]['AvgLatency'] for tps in ordered_tps]
plt.boxplot(latency_data_1, positions=np.arange(len(ordered_tps))-0.25, widths=0.10,
            patch_artist=True, boxprops=dict(facecolor='lightcoral', alpha=0.4),
            medianprops=dict(color='darkred', linewidth=2), showfliers=False)
mean_latency_1 = [np.mean(data) for data in latency_data_1]
plt.plot(np.arange(len(ordered_tps))-0.25, mean_latency_1, linestyle='-', marker='o', color='red', label='reportStatus')

# Boxplot TPS 2 - statusReports
latency_data_2 = [df2[df2['ConfiguredTPS'] == tps]['AvgLatency'] for tps in ordered_tps]
plt.boxplot(latency_data_2, positions=np.arange(len(ordered_tps)), widths=0.10,
            patch_artist=True, boxprops=dict(facecolor='skyblue', alpha=0.6),
            medianprops=dict(color='darkblue', linewidth=2), showfliers=False)
mean_latency_2 = [np.mean(data) if len(data) > 0 else np.nan for data in latency_data_2]
plt.plot(np.arange(len(ordered_tps)), mean_latency_2, linestyle='--', marker='s', color='blue', label='statusReports')

# Boxplot TPS 3 - getLatestStatus
latency_data_3 = [df3[df3['ConfiguredTPS'] == tps]['AvgLatency'] for tps in ordered_tps]
plt.boxplot(latency_data_3, positions=np.arange(len(ordered_tps))+0.25, widths=0.10,
            patch_artist=True, boxprops=dict(facecolor='mediumseagreen', alpha=0.5),
            medianprops=dict(color='darkgreen', linewidth=2), showfliers=False)
mean_latency_3 = [np.mean(data) if len(data) > 0 else np.nan for data in latency_data_3]
plt.plot(np.arange(len(ordered_tps))+0.25, mean_latency_3, linestyle='-.', marker='^', color='green', label='getLatestStatus')

plt.xticks(ticks=np.arange(len(ordered_tps)), labels=ordered_tps)
plt.title('Latncia Mdia por TPS Configurado', fontsize=14)
plt.xlabel('TPS Configurado', fontsize=12)
plt.ylabel('Latncia Mdia (s)', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.4)
plt.legend()
plt.tight_layout()
plt.savefig("grafico_latency_tps.png", dpi=300)
plt.show()

```
